isnull(),and isnull()sum() is used for finding the null values and number of null values in each column(here true indicates null value)

```python
mean_age=df['Age'].mean()
df['Age'].fillna(mean_age,inplace=True)
df
```

```python
df['Cabin'].fillna("B 96",inplace=True)
df
```

```python
df['Embarked'].fillna("S",inplace=True)
df
```

```python
df.rename(columns={'No':'PassengerID'},inplace=True)
df
```

```python
df.isnull().sum()
```

Null values are filled appropriately with specific strings and mean of values.After all the process now our data does not have any null data its proven by isnull().sum()command

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['Survived'],bins=2,edgecolor='b',linewidth=1,width=0.4)
plt.xlabel('Survivers')
plt.ylabel('Frequency')
plt.title('survived graph')
plt.show()
```

```python
plt.plot(df['Age'],df['Survived'])
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show
```

```python
plt.step(df['Age'],df['Survived'])
plt.xlabel('Age')
plt.ylabel('Survived')
plt.show
```

A histogram is plotted on the basis of survived passengers. Most of the passengers did not survived.plot and step graphs of survived vs age is ploted.There is no specific co relation between them.
